#### There was a memory issue when using the entire dataset, so we took a random sample of rows from the Ratings file.

```python
df_2d = df_2a.sample(n=70000)    # random sample n = 70000
```

```python
df_2d.head(20)
```

#### Create a matrix and fill 0 values

```python
book_user_rating_pivot = df_2d.pivot(index = 'User-ID', columns = 'ISBN', values = 'Book-Rating').fillna(0)
book_user_rating_pivot.head(10)
```

```python
book_user_rating_pivot.shape
```

```python
X = book_user_rating_pivot.values.T
X.shape
```

#### Fit the model using TruncatedSVD

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings                                                    # this takes a while to execute
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

#### Check the result by providing one ISBN value into .index()

```python
book_title = book_user_rating_pivot.columns
book_title_list = list(book_title)
coffey_hands = book_title_list.index("0380715899")       # pick one ISBN value from df_2d and insert in .index()

# Note: Some ISBN values show no result, indicating the ISBN is a cold start with no rating!
# So, good to start with ISBN receiving a rating to see the result 
```

```python
corr_coffey_hands  = corr[coffey_hands]
list1 = list(book_title[(corr_coffey_hands >= 0.9)])[:20]         # pick top 20 books with correlation coefficient >0.9
list1
```
